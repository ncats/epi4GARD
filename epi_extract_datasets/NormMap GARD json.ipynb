## GOAL: Save GARD into NormMap GARD Dict JSON

```python
import json, codecs
diseases = json.load(codecs.open('gard-id-name-synonyms.json', 'r', 'utf-8-sig'))
```

```python
diseases
```

```python
#key is GARD ID, value is list of names
GARD_dict = {}
for entry in diseases:
    print(entry)
    dz_names = []
    dz_names.append(entry['name'])
    if entry['synonyms']:
        dz_names+=entry['synonyms']
    GARD_dict[entry['gard_id']] = dz_names
```

```python
GARD_dict
```

```python
with open('./NormMap/GARDictionary4NormMap.json', 'w') as f:
    json.dump(GARD_dict, f)
```
